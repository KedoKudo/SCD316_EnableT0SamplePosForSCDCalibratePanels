```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# profile L1-T0 space

def profileL1T0(filename, frac=0.05):
    df_full = pd.read_csv(filename, delimiter="\t")
    # reduce plotting density
    df = df_full.sample(frac=frac)

    #-- get the global minimum and report on terminal
    global_minimum = df[df["residual"]==df["residual"].min()]
    print(f"Global minimum is at:\n{global_minimum}")
    
    #-- generate figure
    fig = plt.figure(figsize=(10, 20))
    ax = fig.add_subplot(1, 1, 1, projection='3d')

    surf = ax.plot_trisurf(
        df["dL1"],
        df["dT0"],
        df["residual"],
        linewidth=0.1,
        cmap='jet',
    )

    ax.set_title("$\Delta L_1$-$\Delta T_0$ parameter space")
    ax.set_xlabel("$\Delta L_1$ (meter)")
    ax.set_ylabel("$\Delta T_0$ (ms)")
    ax.view_init(elev=+14, azim=-48)
    
#     fig.colorbar(surf, shrink=0.2, aspect=10)
```

```python
# ideal case

profileL1T0("data/profileSCDCalibratePanels2_L1T0_idealTOPAZ.csv", frac=0.05)
```
